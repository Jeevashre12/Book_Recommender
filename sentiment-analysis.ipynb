```python
import numpy as np

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]
isbn = []
emotion_scores = {label: [] for label in emotion_labels}

def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotion_labels}

    for prediction in predictions:
        sorted_predictions = sorted(prediction, key=lambda x: x["label"])

        for index, label in enumerate(emotion_labels):
            per_emotion_scores[label].append(sorted_predictions[index]["score"])

    return {label: np.max(scores) if scores else 0 for label, scores in per_emotion_scores.items()}
```

```python
for i in range(10):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(".")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)

    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
from tqdm import tqdm

emotion_labels = ["anger", "disgust", "fear", "joy", "sadness", "surprise", "neutral"]

isbn = []
emotion_scores = {label: [] for label in emotion_labels}

batch_size = 50
total_books = len(books)

for start in range(0, total_books, batch_size):
    end = min(start + batch_size, total_books)
    batch_books = books.iloc[start:end]

    remaining = total_books - end
    print(f"\nProcessing books {start+1} to {end} | Remaining: {remaining}")

    for _, row in batch_books.iterrows():
        isbn.append(row["isbn13"])

        desc = str(row["description"])
        sentences = [s.strip() for s in desc.split(".") if s.strip()]

        predictions = classifier(sentences, truncation=True)
        max_scores = calculate_max_emotion_scores(predictions)

        for label in emotion_labels:
            emotion_scores[label].append(max_scores[label])
```

```python
emotion_df = pd.DataFrame(emotion_scores)
emotion_df["isbn13"] = isbn
```

```python
emotion_df
```

```python
books=pd.merge(books,emotion_df,on="isbn13")
```
